```python
cat_cols = Insuarance_data.select_dtypes(include=['object']).columns.tolist()
cat_cols
```

```python
cat_data = Insuarance_data[cat_cols]
```

```python
cat_data_dummies = pd.get_dummies(cat_data)
```

```python
cat_data_dummies.shape
```

```python
cat_data_dummies.describe()
```

```python
numeric_cols = Insuarance_data.select_dtypes(include=[np.number]).columns.tolist()
numeric_cols
```

```python
num_data = Insuarance_data[numeric_cols]
```

```python
Insuarance_Data_dummy = pd.concat([num_data, cat_data_dummies], axis=1)
```

```python
Insuarance_Data_dummy.shape
```

```python
cat_test = Insuarance_test_data[cat_cols]
cat_test_dummies = pd.get_dummies(cat_test)
print(cat_test_dummies.shape)
num_test = Insuarance_test_data[numeric_cols]
Insuarance_test_dummy = pd.concat([num_test, cat_test_dummies], axis=1)
print(Insuarance_test_dummy.shape)
```

```python
y = Insuarance_Data_dummy['CarInsurance']
Insuarance_Data_dummy.drop('CarInsurance',axis=1,inplace=True)
Insuarance_test_dummy.drop('CarInsurance',axis=1,inplace=True)
print(y.shape)
print(Insuarance_Data_dummy.shape)
print(Insuarance_test_dummy.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(Insuarance_Data_dummy, y, train_size=0.7 , random_state=100)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score,roc_auc_score
accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred)
```

```python
print('Accuracy is  :' ,accuracy)
print('F1 score is :' ,f1)
print('Precision is :',precision)
print('Recall is :',recall)
print('Roc Auc is :',roc_auc)
```
